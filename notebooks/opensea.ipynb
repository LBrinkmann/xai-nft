```python
# Parameters
top_folder = "../data/top_collections"
n_days = 365
collections = ["boredapeyachtclub"]
```

```python
%load_ext autoreload
%autoreload 2

import os
import pandas as pd
from xai.utils.time import create_timeline
from xai.utils.data import save_json, check_file, load_json
from xai.api.opensea import get_collection, get_collection_assets, get_all_events
```

```python
# get details for top collections
folder = 'details'
for collection in collections:
    filename = os.path.join(top_folder, folder, f'{collection}.json')
    if check_file(filename):
        print(f"Skip collection {collection}.")
    else:
        try:
            collection_details = get_collection(collection=collection)
            save_json(collection_details, filename)
            print(f"Saved collection {collection}.")
        except Exception as e:
            raise e
            print(e)
            print(f"Failed with collection {collection}.")
```

```python
# get assets for top collections
folder = 'assets'
for collection in collections:
    filename = os.path.join(top_folder, folder, f'{collection}.json')
    if check_file(filename):
        print(f"Skip collection {collection}.")
    else:
        try:
            collection_details = get_collection_assets(collection=collection)
            save_json(collection_details, filename)
            print(f"Saved collection {collection}.")
        except:
            print(f"Failed with collection {collection}.")
```

```python
# get events for top collections
tl = create_timeline(n_days)

for collection in collections:
    filename = os.path.join(top_folder, 'events', f'{collection}_today.json')
    events = get_all_events(
        collection=collection, before=None, after=tl[0]['before'])
    save_json(events, filename)
    for t in tl:
        filename = os.path.join(top_folder, 'events', f"{collection}_{t['day']}.json")
        if check_file(filename):
            print(f"Skip collection {collection} for {t['day']}.")
        else:
            try:
                events = get_all_events(
                    collection=collection, before=t['before'], after=t['after'])
                save_json(events, filename)
                print(f"Saved collection {collection} for {t['day']}.")
            except:
                print(f"Failed with collection {collection} for {t['day']}.")
```
